```python
'''计算每个用户的度数'''

degrees = {user: len(info['friends']) for user, info in graph.items()}  
sorted_degrees = sorted(degrees.items(), key=lambda x: x[1], reverse=True)  
top_5_degrees = sorted_degrees[:5]  
print("度数最多的前5个用户及其朋友数量:", top_5_degrees)
```

```python
'''找出影响力最高的前五个用户'''

sorted_influence = sorted(nodes, key=lambda x: x[1], reverse=True)  
top_5_influence = sorted_influence[:5]  
print("影响力最高的前5个用户:", top_5_influence)
```

```python
# 假设影响力分数和朋友数量各占50%的权重  
combined_scores = {user: 0.5 * info['influence_score'] + 0.5 * degrees[user] for user, info in graph.items()}  
sorted_combined = sorted(combined_scores.items(), key=lambda x: x[1], reverse=True)  
top_influential_users = sorted_combined[:5]  
print("对社交网络最有影响力的用户:", top_influential_users)
```

```python
'''找出 Yanyan-Wang 的“朋友的朋友”中最有可能成为新朋友的用户'''
def find_common_friends(user1, user2, graph):  
    return set(graph[user1]['friends'].keys()).intersection(set(graph[user2]['friends'].keys()))  
  
yanyan_wang_friends = graph["Yanyan-Wang"]['friends']  
recommendations = {}  
  
for friend in yanyan_wang_friends:  
    friend_friends = graph[friend]['friends']  
    for other_friend in friend_friends:  
        if other_friend != "Yanyan-Wang" and other_friend not in yanyan_wang_friends:  
            if other_friend not in recommendations:  
                recommendations[other_friend] = 0  
            recommendations[other_friend] += 1  
  
sorted_recommendations = sorted(recommendations.items(), key=lambda x: x[1], reverse=True)  
print("推荐给 Yanyan-Wang 的新朋友（最多共同朋友）:", sorted_recommendations)
```

```python
'''找出与 Yanyan-Wang 连接强度最高的5个用户'''
# 这里假设连接强度就是边的权重之和  
yanyan_wang_connections = graph["Yanyan-Wang"]['friends']  
connection_strengths = {friend: yanyan_wang_connections[friend] for friend in yanyan_wang_connections}  
sorted_strengths = sorted(connection_strengths.items(), key=lambda x: x[1], reverse=True)  
top_5_connections = sorted_strengths[:5]  
print("与 Yanyan-Wang 连接强度最高的5个用户:", top_5_connections)
```

```python
'''  使用图论可视化工具绘制用户关系网络图'''
import networkx as nx  
import matplotlib.pyplot as plt  
import matplotlib.colors as mcolors

# 创建图对象  
G = nx.Graph()  
  
# 添加节点和边  
for node in nodes:  
    G.add_node(node[0], influence_score=node[1])  
for edge in edges:  
    user1, user2, weight = edge  
    G.add_edge(user1, user2, weight=weight)  
  
# 绘制图形  
pos = nx.spring_layout(G)  # 使用spring布局  
# 设置节点大小、颜色和标签颜色  
node_sizes = [info['influence_score'] * 5 for _, info in G.nodes(data=True)]  # 调整节点大小比例  
# 使用红色系的颜色映射，如从浅红到深红  
vmin, vmax = min(node[1] for node in nodes), max(node[1] for node in nodes)  
norm = mcolors.Normalize(vmin=vmin, vmax=vmax)  
cmap = plt.cm.Reds  # 使用Reds映射，使低值对应浅红色调  
node_colors = [cmap(norm(info['influence_score'])) for _, info in G.nodes(data=True)]  
  
# 绘制图，设置边颜色为灰色或更淡的颜色  
plt.figure(figsize=(10, 8))  
nx.draw(G, pos, with_labels=False, node_size=node_sizes, node_color=node_colors, edge_color='lightgray')  
# 单独绘制标签，并设置标签颜色为更浅的颜色，如浅灰色或淡粉色  
label_color = 'lightpink'  # 使用浅灰色作为标签颜色  
# 或者使用淡粉色：label_color = '#FFB6C1'  
labels = {node: node for node, _ in G.nodes(data=True)}  # 使用节点名作为标签  
nx.draw_networkx_labels(G, pos, labels, font_size=12, font_color=label_color)  
  
plt.title('Graph with Influence Scores')  
plt.show()
```
