```python
import pandas as pd

# 读取数据集
df = pd.read_csv(r"C:\Users\86178\Downloads\github_bot_processed_data.csv")

# 格式化日期列
# 假设 'created_at' 和 'updated_at' 列是日期时间字符串，格式为 '%Y-%m-%d %H:%M:%S%z'
df['created_at'] = pd.to_datetime(df['created_at'], format='%Y-%m-%d %H:%M:%S%z')
df['updated_at'] = pd.to_datetime(df['updated_at'], format='%Y-%m-%d %H:%M:%S%z')

# 将日期格式化为所需的字符串格式，例如 'YYYY-MM-DD'
df['created_at_formatted'] = df['created_at'].dt.strftime('%Y-%m-%d')
df['updated_at_formatted'] = df['updated_at'].dt.strftime('%Y-%m-%d')

# 使用 '{:,.0f}' 格式化字符串来添加千位分隔符
df['public_repos_formatted'] = df['public_repos'].apply(lambda x: f'{x:,.0f}' if pd.notna(x) else '')
df['followers_formatted'] = df['followers'].apply(lambda x: f'{x:,.0f}' if pd.notna(x) else '')

# 由于数据集中没有 'profit_margin'，我们仅展示已格式化的列
print(df[['label', 'created_at_formatted', 'updated_at_formatted', 'public_repos_formatted', 'followers_formatted']])
```

```python
# -*- coding: utf-8 -*-
"""
Created on Thu Nov 21 13:15:11 2024

@author: 86178
"""

import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  

# 读取数据集
df = pd.read_csv(r"C:\Users\86178\Downloads\github_bot_processed_data.csv")

# 选择需要进行对数变换的列  
columns_to_log_transform = ['public_repos', 'public_gists', 'followers', 'following']  

# 应用对数变换  
for col in columns_to_log_transform:  
    df[col + '_log'] = np.log1p(df[col])  

# 打印原始和变换后的数据对比  
print(df[columns_to_log_transform + [col + '_log' for col in columns_to_log_transform]])  

# 可视化对数变换的影响  
plt.figure(figsize=(14, 8))  

# 原始数据的分布  
plt.subplot(2, 2, 1)  
plt.hist(df['public_repos'], bins=30, color='blue', alpha=0.7, edgecolor='black')  
plt.title('Distribution of Public Repos (Original)')  
plt.xlabel('Public Repos')  
plt.ylabel('Frequency')  

plt.subplot(2, 2, 2)  
plt.hist(df['public_gists'], bins=30, color='blue', alpha=0.7, edgecolor='black')  
plt.title('Distribution of Public Gists (Original)')  
plt.xlabel('Public Gists')  
plt.ylabel('Frequency')  

# 对数变换后的数据分布  
plt.subplot(2, 2, 3)  
plt.hist(df['public_repos_log'], bins=30, color='green', alpha=0.7, edgecolor='black')  
plt.title('Distribution of Public Repos (Log Transformed)')  
plt.xlabel('Public Repos (Log)')  
plt.ylabel('Frequency')  

plt.subplot(2, 2, 4)  
plt.hist(df['public_gists_log'], bins=30, color='green', alpha=0.7, edgecolor='black')  
plt.title('Distribution of Public Gists (Log Transformed)')  
plt.xlabel('Public Gists (Log)')  
plt.ylabel('Frequency')  

plt.tight_layout()  
plt.show()
```

```python
import pandas as pd  
import matplotlib.pyplot as plt  

# 读取数据集  
df = pd.read_csv(r"C:\Users\86178\Downloads\github_bot_processed_data.csv")

# 选择布尔特征  
bool_features = ['site_admin', 'company', 'blog', 'location', 'hireable']  

# 计算每个布尔特征的统计数据  
counts = df[bool_features].apply(lambda x: x.value_counts(), axis=0).fillna(0)  

# 转置数据以方便绘图  
counts = counts.T  

# 设置图表大小  
plt.figure(figsize=(10, 6))  

# 绘制堆积柱状图  
bar_width = 0.35  
index = range(len(bool_features))  

# 绘制True和False的条形  
plt.bar(index, counts[True], bar_width, label='True', color='skyblue')  
plt.bar(index, counts[False], bar_width, bottom=counts[True], label='False', color='lightcoral')  

# 设置标题和坐标轴  
plt.title('Distribution of Boolean Features')  
plt.xlabel('Boolean Features')  
plt.ylabel('Count')  
plt.xticks(index, bool_features)  # 设置x轴刻度名称  
plt.legend()  

# 显示每个柱子的数字  
for i in index:  
    plt.text(i, counts[True][i] + counts[False][i] / 2, str(int(counts[False][i])), ha='center', va='center')  
    plt.text(i, counts[True][i] + 0.5, str(int(counts[True][i])), ha='center', va='bottom')  

# 显示图表  
plt.tight_layout()  
plt.show()
```

```python
import pandas as pd  
import numpy as np  
import seaborn as sns  
import matplotlib.pyplot as plt  

# 读取 CSV 文件  
df = pd.read_csv(r"C:\Users\86178\Downloads\github_bot_processed_data.csv") 

# 计算各个特征的对数值  
df['log_public_repos'] = np.log1p(df['public_repos'])  
df['log_public_gists'] = np.log1p(df['public_gists'])  
df['log_followers'] = np.log1p(df['followers'])  
df['log_following'] = np.log1p(df['following'])  

# 选择要计算相关性的特征  
correlation_features = ['log_public_repos', 'log_public_gists', 'log_followers', 'log_following']  

# 计算相关性矩阵  
correlation_matrix = df[correlation_features].corr()  

# 设置绘图风格  
sns.set(style='white')  

# 创建热图  
plt.figure(figsize=(8, 6))  
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})  

# 添加标题  
plt.title('Correlation Heatmap of Log Features')  

# 显示图形  
plt.show()
```
